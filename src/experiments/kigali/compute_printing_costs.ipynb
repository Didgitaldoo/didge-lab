```python
import numpy as np

def compute_wall_volume(bore_profile, wall_thickness, steps=1000):
    # Extract bore length and diameters
    z0, d0 = bore_profile[0]
    z1, d1 = bore_profile[1]
    length = z1 - z0

    # Setup z-axis sampling points
    zs = np.linspace(z0, z1, steps)
    dz = length / (steps - 1)

    # Linear interpolation of diameters
    inner_diams = d0 + (d1 - d0) * (zs - z0) / (z1 - z0)
    r_inners = inner_diams / 2
    r_outers = r_inners + wall_thickness

    # Volume differential at each segment
    dVs = np.pi * (r_outers**2 - r_inners**2) * dz

    # Total volume in mm続
    total_volume_mm3 = np.sum(dVs)

    # Convert to cm続 or mL
    total_volume_cm3 = total_volume_mm3 / 1000

    return total_volume_cm3

# Example usage
bore = [[0, 32], [100, 42]]  # length in mm, diameters in mm
wall_thickness = 4  # mm

volume = compute_wall_volume(bore, wall_thickness)
print(f"Wall volume: {volume:.2f} cm続")

density_pla = 1.24 # g/cm3
weight = volume*density_pla
print(f"Weight: {weight:.2f}g")
```
