```python
import math

scale = [0, 2, 3, 5, 7, 8, 11]
scale_key = -31

scale_freqs = []
freq = -1
i=0
while freq<1000:
    f1 = math.floor(i/len(scale))
    f2 = i%len(scale)
    note = scale_key + 12*f1 + scale[f2]
    freq = note_to_freq(note)
    i += 1
    scale_freqs.append(freq)

scale_freqs = np.log2(scale_freqs)
scale_freqs
```

```python
freqs = np.random.choice(scale_freqs, 5, replace=False)
freqs += np.random.sample(len(freqs)) - 0.5
freqs
```

```python
i = [np.argmin(np.abs(scale_freqs-freq)) for freq in freqs]
diffs = np.power(freqs - scale_freqs[i], 2)
diffs
```
