```python
def diff(geo1, geo2):
    if geo1.length()>geo2.length():
        x=geo1
        geo1=geo2
        geo2=x

    x = np.arange(0, geo1.length(), geo1.length()/100)
    y1 = np.array([geo1.diameter_at_x(_x) for _x in x])
    y2 = np.array([geo2.diameter_at_x(_x) for _x in x])

    diff = np.abs(y1-y2).sum()

    lf = geo1.length()/geo2.length()
    if lf<0:
        lf = 1/lf
    diff *= lf

    return diff
              
geo1 = Geo(population[0]["representation"]["geo"])

df = []
for i in range(len(population)):
    geo2 = Geo(population[i]["representation"]["geo"])
    df.append({"i": i, "diff": diff(geo1, geo2)})

df = pd.DataFrame(df)
df
```
