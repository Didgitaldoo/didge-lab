```python
import pickle
import sys
sys.path.append('../../')
from tairua import visualize_population, TairuaLoss
import pickle
import numpy as np
from didgelab.calc.sim.sim import quick_analysis
from didgelab.util.didge_visualizer import plot_geo_to_axs
import matplotlib.pyplot as plt


pop = pickle.load(open("results/pop1.bin", "rb"))
target_freqs, target_impedances = pickle.load(open("results/targets.bin", "rb"))

population = []
limit = 20
for i in pop:
    geo = i.genome2geo()
    analysis = quick_analysis(geo, log_freqs=False)
    analysis["notes"] = json.loads(analysis["notes"].to_json())
    analysis["impedance"] = analysis["impedance"] / analysis["impedance"].max()
    population.append({
        "genome": i.genome.tolist(),
        "geo": geo.geo,
        "loss": i.loss,
        "analysis": analysis
    })
    if len(population) >= limit:
        break

data = {
    "population": population,
    "target": {
        "target_freqs": target_freqs,
        "target_impedances": target_impedances
    },
    "evolution_progress": {
        "generation_size": 100, 
        "num_generations": 100,  
        "population_size": 1000, 
        "current_generation": 125,
        "elapsed_time": "23min",
        "est. remaining time": "73min",
        "time per generation": "76sec"
    }
}

outfolder = "../../../../didgelab-web/example_shapes/"
outfile = outfolder + "tairua_pop.json"

import json
from didgelab.evo.nuevolution import NumpyEncoder
encoder = NumpyEncoder()
data = encoder.encode(data)
with open(outfile, "w") as f:
    f.write(data)
# json.dump(data, open(outfile, "w"))
```
