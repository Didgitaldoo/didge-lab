# High Inharmonic Loss

**Purpose:** Maximize **inharmonicity**: push harmonic ratios away from integers for a metallic, dissonant timbre.

**Formula:**

$$L_{inharm} = w \cdot \left( 0.5 - \text{avg}\left(\left| \frac{f_n}{f_0} - \lfloor \frac{f_n}{f_0} \rceil \right|\right) \right)$$

**Symbols:**
- $ \lfloor x \rceil $: Nearest integer (rounding).
- $ f_n/f_0 $: Harmonic ratio.
- $ 0.5 $: Max possible deviation from an integer ratio.
- $ w $: Weight for metallic/chaotic tone.
