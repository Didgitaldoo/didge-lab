# Stretched Odd Harmonic Loss

**Purpose:** Emphasize **odd** harmonics (1st, 3rd, 5th) with slight stretching for a hollow/woody timbre.

**Formula:**

$$L_{odd} = \frac{w}{600} \cdot \sum \left| 1200 \cdot \log_2 \left( \frac{f_{closest}}{f_{target}} \right) \right|$$

Targets are $ f_0 $, $ 3.1\,f_0 $, $ 5.2\,f_0 $ (odd-like, slightly stretched). 600 normalizes cent-based error.
